# Mapas de Calor (Heatmaps)  
Importamos Seaborn  
Cargamos el dataset "Flights" que viene con Seaborn

```python
import seaborn as sns
vuelos= sns.load_dataset('flights')
```

```python
vuelos.head(10)
```

Para crear el heatmap debemos tener la tabla en formato de matriz

```python
vuelos_matriz = vuelos.pivot_table(index='month', columns='year', values='passengers')
vuelos_matriz
```

```python
sns.heatmap(vuelos_matriz)
```

Mediante un nuevo atributo puedo cambiar el formato de los colores.  
[Matplotlib Colormap Reference](https://matplotlib.org/stable/gallery/color/colormap_reference.html)

```python
sns.heatmap(vuelos_matriz, cmap='viridis')
```

```python
sns.heatmap(vuelos_matriz, cmap='rainbow')
```

```python
sns.heatmap(vuelos_matriz, cmap='hsv', linecolor='white', linewidths=1)
```

# Gr√°ficos de cuadr√≠culas (Grids). PairGrid y FacetGrid 

Importamos Matplotlib

```python
import matplotlib.pyplot as plt
```

Cargamos un dataset de flores que pertenece a Seaborn

```python
flores = sns.load_dataset('iris')
flores.head(10)
```

```python
sns.pairplot(flores)
```

**PairGrid**

```python
graficos = sns.pairplot(flores)
graficos.map(plt.scatter)
```

```python
graficos = sns.pairplot(flores)
graficos.map_diag(sns.histplot)
graficos.map_upper(sns.regplot)
graficos.map_lower(sns.kdeplot)
```

**FacetGrid**

```python
propinas = sns.load_dataset('tips')
propinas.head(10)
```

```python
propinas['time'].unique()
```

```python
graficos2 = sns.FacetGrid(data=propinas, col='time', row='smoker')
graficos2.map(sns.histplot, 'total_bill')
```

```python
graficos2 = sns.FacetGrid(data=propinas, col='time', row='smoker')
graficos2.map(sns.scatterplot, 'total_bill', 'tip')
```

```python
graficos2 = sns.FacetGrid(data=propinas, col='time', row='smoker')
graficos2.map(plt.scatter, 'total_bill', 'tip')
```
