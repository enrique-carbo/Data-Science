### Importo Seaborn   
### Cargo un dataset llamado "tips"

```python
import seaborn as sns
propinas = sns.load_dataset('tips')
propinas.head(10)
```

## **Gráficos de Distribución**
### Gráfico de Histogramas: estos evalúan una columna o una variable

```python
sns.distplot(propinas['total_bill'])
```

```python
sns.distplot(propinas['total_bill'], bins=30)
```

```python
sns.displot(propinas['total_bill'])
```

```python
sns.displot(propinas['tip'])
```

```python
sns.displot(propinas['total_bill'], bins=50)
```

```python
sns.histplot(propinas['total_bill'])
```

### Para comparar dos variables o columnas utilizamos otro tipo de gráfico "jointplot"

```python
sns.jointplot(x='total_bill', y='tip', data=propinas)
```

```python
sns.jointplot(x='total_bill', y='tip', data=propinas, kind='hex')
```

### Gráfico de Regresión Lineal

```python
sns.jointplot(x='total_bill', y='tip', data=propinas, kind='reg')
```

### Gráfico de Densidad: cuanto mayor es el color mayor es la densidad

```python
sns.jointplot(x='total_bill', y='tip', data=propinas, kind='kde', fill=True)
```

### Para comparar todas las variables numéricas existe el tipo de gráfico "pairplot"

```python
sns.pairplot(propinas)
```

```python
sns.pairplot(propinas, hue='sex')
```

```python
sns.pairplot(propinas, hue='size')
```

```python
sns.rugplot(propinas['total_bill'])
```

## Gráficos para columnas tipo categorias

```python
import numpy as np

propinas.head(10)
```

```python
sns.barplot(x='sex', y='total_bill', data=propinas)
```

```python
sns.countplot(x='sex', data=propinas)
```

```python
sns.countplot(x='smoker', data=propinas)
```

### Diagrama de cajas

```python
sns.boxplot(x='day', y='total_bill', data=propinas)
```

```python
sns.boxplot(x='day', y='total_bill', data=propinas, hue='smoker')
```

```python
sns.violinplot(x='day', y='total_bill', data=propinas)
```

```python
sns.violinplot(x='day', y='total_bill', data=propinas, hue='smoker', split=True)
```

```python
sns.stripplot(x='day', y='total_bill', data=propinas, hue='sex')
```

```python
sns.swarmplot(x='day', y='total_bill', data=propinas, hue='sex')
```
