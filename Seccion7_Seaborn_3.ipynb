# Gráficos de Regresión. Regression Plots

```python
import seaborn as sns
propinas = sns.load_dataset('tips')
propinas.head(10)
```

```python
sns.lmplot(x='total_bill', y='tip', data=propinas)
```

```python
sns.lmplot(x='total_bill', y='tip', data=propinas, hue='sex')
```

```python
sns.lmplot(x='total_bill', y='tip', data=propinas, hue='sex', markers=['o', 'v'])
```

```python
sns.lmplot(x='total_bill', y='tip', data=propinas, hue='sex', markers=['o', 'v'], scatter_kws={'s': 90})
```

```python
sns.lmplot(x='tip', y='total_bill', data=propinas, col='sex')
```

```python
sns.lmplot(x='tip', y='total_bill', data=propinas, col='day')
```

```python
sns.lmplot(x='tip', y='total_bill', data=propinas, col='day', row='sex')
```

```python
sns.lmplot(x='tip', y='total_bill', data=propinas, col='day', hue='sex')
```

```python
sns.lmplot(x='total_bill', y='tip', data=propinas, aspect=1.2, height=10, hue='sex')
```

## Estilos y colores en Seaborn

```python
import matplotlib.pyplot as plt
propinas.head(10)
```

```python
sns.set_style('darkgrid')
sns.countplot(x='sex', data=propinas)
```

```python
sns.set_context('notebook', font_scale=1.3)
plt.figure(figsize=(8,10))
sns.set_style('white')
sns.countplot(x='sex', data=propinas)
sns.despine()
```

### Como cambiar los colores  
[Seaborn Color Palette](https://seaborn.pydata.org/tutorial/color_palettes.html?highlight=palette)

```python
sns.countplot(x='sex', data=propinas, palette='rocket')
```

```python
colores = ['#15A009', 'blue']
sns.countplot(x='sex', data=propinas, palette=colores)
```
